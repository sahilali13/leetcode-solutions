## Question 550 -> 575

    Difficulty - Hard
    Algo - Dynamic Programming

An attendance record for a student can be represented as a string where each character signifies whether the student was absent, late, or present on that day. The record only contains the following three characters:

-   `'A'`: Absent.
-   `'L'`: Late.
-   `'P'`: Present.

Any student is eligible for an attendance award if they meet **both** of the following criteria:

-   The student was absent (`'A'`) for **strictly** fewer than 2 days **total**.
-   The student was **never** late (`'L'`) for 3 or more **consecutive** days.

Given an integer `n`, return _the **number** of possible attendance records of length_ `n` _that make a student eligible for an attendance award. The answer may be very large, so return it_ **modulo** <code>10<sup>9</sup> + 7</code>.

**Constraints:**

-   <code>1 <= n <= 10<sup>5</sup></code>


```python
class Solution:
    def checkRecord(self, n: int) -> int:
        MOD = 10**9 + 7

        # Initialize the DP arrays to handle sub-problems.
        dp_curr = [[0] * 3 for _ in range(2)]
        dp_next = [[0] * 3 for _ in range(2)]

        # Base case: one way to have a record of length 0.
        dp_curr[0][0] = 1

        for _ in range(n):
            # Reset dp_next for the new iteration.
            dp_next = [[0] * 3 for _ in range(2)]

            for total_absences in range(2):
                for consecutive_lates in range(3):
                    count = dp_curr[total_absences][consecutive_lates]

                    # Case when 'P' is chosen.
                    dp_next[total_absences][0] = (
                        dp_next[total_absences][0] + count
                    ) % MOD

                    # Case when 'A' is chosen.
                    if total_absences < 1:
                        dp_next[total_absences + 1][0] = (
                            dp_next[total_absences + 1][0] + count
                        ) % MOD

                    # Case when 'L' is chosen.
                    if consecutive_lates < 2:
                        dp_next[total_absences][consecutive_lates + 1] = (
                            dp_next[total_absences][consecutive_lates + 1] + count
                        ) % MOD

            # Update dp_curr to the results of dp_next for the next iteration.
            dp_curr = dp_next

        # Sum up the counts for all valid records of length n.
        result = (
            sum(
                dp_curr[total_absences][consecutive_lates]
                for total_absences in range(2)
                for consecutive_lates in range(3)
            )
            % MOD
        )

        return result


if __name__ == "__main__":
    sol = Solution()
    cases = [{"n": 2}, {"n": 1}, {"n": 10101}]
    for case in cases:
        print(sol.checkRecord(case["n"]))
```
