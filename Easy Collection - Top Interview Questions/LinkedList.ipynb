## Delete Node in a Linked List

    There is a singly-linked list head and we want to delete a node node in it.

    You are given the node to be deleted node. You will not be given access to the first node of head.

    All the values of the linked list are unique, and it is guaranteed that the given node node is not the last node in the linked list.

    Delete the given node. Note that by deleting the node, we do not mean removing it from memory. We mean:

    * The value of the given node should not exist in the linked list.
    * The number of nodes in the linked list should decrease by one.
    * All the values before node should be in the same order.
    * All the values after node should be in the same order.


```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x: int) -> None:
        self.val = x
        self.next = None
    
class LinkedList:
    def __init__(self) -> None:
        self.head = None
        self.head = None
        self.length = 0
    
    def __str__(self) -> str:
        result = ''
        if self.length != 0:
            current = self.head
            while current:
                result += str(current.val) + ' '
                current = current.next
        return result

    def add(self, node: ListNode) -> None:
        if self.length == 0:
            self.head = node
        else:
            self.tail.next = node
        self.tail = node
        self.length += 1

class Solution:
    def deleteNode(self, node):
        """
        :type node: ListNode
        :rtype: void Do not return anything, modify node in-place instead.
        """
        temp = node.next
        node.val = temp.val
        node.next = temp.next
        del temp

if __name__ == '__main__':
    sol = Solution()

    cases = [([4,5,1,9], 5),
            ([4,5,1,9], 1)]
    
    for case in cases:
        temp = LinkedList()
        for val in case[0]:
            temp.add(node = ListNode(x = val))
        current = temp.head
        while current:
            if current.val == case[1]:
                node = current
                break
            current = current.next
        sol.deleteNode(node = current)

        print(temp, case[1])
            
    
        
```

## Remove Nth Node From End of List

    Given the head of a linked list, remove the nth node from the end of the list and return its head.


```python
# Definition for singly-linked list.
class ListNode: ...
class ListNode:
    def __init__(self, val: int = 0, next: ListNode = None) -> None:
        self.val = val
        self.next = next
    
    def __str__(self) -> str:
        return str(self.val)
    
class LinkedList:
    def __init__(self) -> None:
        self.head = None
        self.tail = None
        self.length = 0
    
    def __str__(self) -> str:
        result = ''
        if self.length != 0:
            current = self.head
            while current:
                result += str(current.val) + ' '
                current = current.next
        return result

    def add(self, node: ListNode) -> None:
        if self.length == 0:
            self.head = node
        else:
            self.tail.next = node
        self.tail = node
        self.length += 1

class Solution:
    def removeNthFromEnd(self, head: ListNode, n: int) -> ListNode:
        count = 0
        prevNode, currentNode = ListNode(next = head), head
        while currentNode:
            if count >= n:
               prevNode = prevNode.next
            currentNode = currentNode.next
            count += 1
        if prevNode.next == head:
            head = head.next
        else:
            prevNode.next = prevNode.next.next
        del count, prevNode, currentNode
        return head
           
if __name__ == '__main__':
    sol = Solution()

    cases = [([1,2,3,4,5], 2),
            ([1], 1),
            ([1,2], 1),
            ([1,2], 2)]
    
    for case in cases:
        temp = LinkedList()
        for val in case[0]:
            temp.add(node = ListNode(val = val))
        ans = sol.removeNthFromEnd(head = temp.head, n = case[1])
        print(ans)
        
```

## Reverse Linked List

    Given the head of a singly linked list, reverse the list, and return the reversed list.


```python
# Definition for singly-linked list.
class ListNode: ...
class ListNode:
    def __init__(self, val: int = 0, next: ListNode = None) -> None:
        self.val = val
        self.next = next
    
    def __str__(self) -> str:
        return str(self.val)
    
class LinkedList:
    def __init__(self) -> ListNode:
        self.head = None
        self.tail = None
        self.length = 0
    
    def __str__(self) -> str:
        result = ''
        if self.length != 0:
            current = self.head
            while current:
                result += str(current.val) + ' '
                current = current.next
        else:
            result = 'None'
        return result 

    def add(self, node: ListNode) -> None:
        if self.length == 0:
            self.head = node
        else:
            self.tail.next = node
        self.tail = node
        self.length += 1

class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        prevNode = None
        curNode = head
        while curNode:
            nextNode = curNode.next
            curNode.next = prevNode
            prevNode = curNode
            curNode = nextNode
        return prevNode
            
if __name__ == '__main__':
    sol = Solution()

    cases = [[1,2],
            [],
            [1]]
    
    for case in cases:
        temp = LinkedList()
        for val in case:
            temp.add(node = ListNode(val = val))
        new_head = sol.reverseList(head = temp.head)
        ans = LinkedList()
        ans.add(new_head)
        print(ans)
        
```

## Merge Two Sorted Lists

    You are given the heads of two sorted linked lists list1 and list2.

    Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

    Return the head of the merged linked list.


```python
# Definition for singly-linked list.
class ListNode: ...
class LinkedList:...

class ListNode:
    def __init__(self, val: int = 0, next: ListNode = None) -> None:
        self.val = val
        self.next = next
    
    def __str__(self) -> str:
        return str(self.val)
    
class LinkedList:
    def __init__(self) -> LinkedList:
        self.head = None
        self.tail = None
        self.length = 0
    
    def __str__(self) -> str:
        result = ''
        if self.length != 0:
            current = self.head
            while current:
                result += str(current.val) + ' '
                current = current.next
        else:
            result = 'None'
        return result 

    def add(self, node: ListNode) -> None:
        if self.length == 0:
            self.head = node
        else:
            self.tail.next = node
        self.tail = node
        self.length += 1

class Solution:
    def mergeTwoLists(self, list1: ListNode, list2: ListNode) -> ListNode | None:
        if list1 and list2:
            if list1.val == list2.val:
                nextNode = list1.next
                list1.next = list2
                list2.next = self.mergeTwoLists(list1 = nextNode, list2 = list2.next)
                return list1
            elif list1.val < list2.val:
                list1.next = self.mergeTwoLists(list1 = list1.next, list2 = list2)
                return list1
            else:
                list2.next = self.mergeTwoLists(list1 = list1, list2 = list2.next)
                return list2
        elif list1 is None and list2:
            return list2
        elif list2 is None and list1:
            return list1
        else:
            return None
            
if __name__ == '__main__':
    sol = Solution()

    cases = [([1,2,4], [1,3,4]),
            ([], []),
            ([], [1])]
    
    for case in cases:
        temp1, temp2 = LinkedList(), LinkedList()
        for val in case[0]:
            temp1.add(node = ListNode(val = val))
        for val in case[1]:
            temp2.add(node = ListNode(val = val))
        print(temp1, end = '\t')
        print(temp2, end = '\t')
        ans = LinkedList()
        ans.add(sol.mergeTwoLists(list1 = temp1.head, list2 = temp2.head))
        print(ans)
        
```

## Palindrome Linked List

    Given the head of a singly linked list, return true if it is a palindrome or false otherwise.


```python
# Definition for singly-linked list.
class ListNode: ...
class LinkedList:...

class ListNode:
    def __init__(self, val: int = 0, next: ListNode = None) -> None:
        self.val = val
        self.next = next
    
    def __str__(self) -> str:
        return str(self.val)
    
class LinkedList:
    def __init__(self) -> LinkedList:
        self.head = None
        self.tail = None
        self.length = 0
    
    def __str__(self) -> str:
        result = ''
        if self.length != 0:
            current = self.head
            while current:
                result += str(current.val) + ' '
                current = current.next
        else:
            result = 'None'
        return result 

    def add(self, node: ListNode) -> None:
        if self.length == 0:
            self.head = node
        else:
            self.tail.next = node
        self.tail = node
        self.length += 1

class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        prevNode = None
        curNode = head
        while curNode:
            nextNode = curNode.next
            curNode.next = prevNode
            prevNode = curNode
            curNode = nextNode
        return prevNode

    def isPalindrome(self, head: ListNode) -> bool:
        if head.next is None:
            return True
        length = 0
        curNode = head
        while curNode:
            length += 1
            curNode = curNode.next
        isEven = True if length % 2 == 0 else False
        revList = head
        counter = 0
        if isEven:
            while counter < length // 2:
                revList = revList.next
                counter += 1
            revList = self.reverseList(head = revList)
            counter = 0
            while counter < length // 2:
                if head.val != revList.val:
                    return False
                counter += 1
                head = head.next
                revList = revList.next
            return True
        else:
            while counter < length // 2 + 1:
                revList = revList.next
                counter += 1
            revList = self.reverseList(head = revList)
            counter = 0
            while counter < length // 2:
                if head.val != revList.val:
                    return False
                counter += 1
                head = head.next
                revList = revList.next
            return True
            
if __name__ == '__main__':
    sol = Solution()

    cases = [[1,2,2,1],
            [1,2],
            [1],
            [1,3,0,2],
            [1,2,2,3,2,2,1],
            [1,1,2,1]]
    
    for case in cases:
        temp = LinkedList()
        for val in case:
            temp.add(node = ListNode(val = val))
        print(temp, end = '\t')
        print(sol.isPalindrome(head = temp.head))
        
        
        
```

## Linked List Cycle

    Given head, the head of a linked list, determine if the linked list has a cycle in it.

    There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the next pointer. Internally, pos is used to denote the index of the node that tail's next pointer is connected to. Note that pos is not passed as a parameter.

    Return true if there is a cycle in the linked list. Otherwise, return false.


```python
# Definition for singly-linked list.
class ListNode: ...
class LinkedList:...

class ListNode:
    def __init__(self, val: int = 0, next: ListNode = None) -> None:
        self.val = val
        self.next = next
    
    def __str__(self) -> str:
        return str(self.val)
    
class LinkedList:
    def __init__(self) -> LinkedList:
        self.head = None
        self.tail = None
        self.length = 0
    
    def __str__(self) -> str:
        result = ''
        if self.length != 0:
            current = self.head
            while current:
                result += str(current.val) + ' '
                current = current.next
        else:
            result = 'None'
        return result 

    def add(self, node: ListNode) -> None:
        if self.length == 0:
            self.head = node
        else:
            self.tail.next = node
        self.tail = node
        self.length += 1

class Solution:
    def hasCycle(self, head: ListNode) -> bool:
        hare = head
        tortoise = head

        while hare is not None and hare.next is not None:
            tortoise = tortoise.next
            hare = hare.next.next

            if tortoise == hare:
                return True
        
        return False

if __name__ == '__main__':
    sol = Solution()
        
        
        
```
